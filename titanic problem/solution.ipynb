```python
train_set = pd.read_csv('./train.csv')
test_set = pd.read_csv('./test.csv')
```

```python
train_set.head()
```

```python
test_set.head()
```

```python
independent_var = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare' ]

train_X = train_set[independent_var].copy()
test_X = test_set[independent_var].copy() 

train_y = train_set['Survived']
```

```python
train_X.head()
```

```python
test_X.head()
```

```python
train_X['Sex'] =  train_X['Sex'].map({'female': 0, 'male': 1})
test_X['Sex'] =  test_X['Sex'].map({'female': 0, 'male': 1})
```

```python
train_X.head()
```

```python
test_X.head()
```

```python
print(np.isnan(train_X).sum())  # Should be 0
print(np.isnan(train_y).sum())  # Should be 0

print(np.isinf(train_X).sum())  # Should be 0
print(np.isinf(train_y).sum())  # Should be 0
```

```python
train_X = train_X.fillna(0) # fill nan values with zeros

print(np.isnan(train_X).sum())  # Should be 0
print(np.isnan(train_y).sum())  # Should be 0

print(np.isinf(train_X).sum())  # Should be 0
print(np.isinf(train_y).sum())  # Should be 0
```

```python
train_X = train_X.to_numpy()
train_y = train_y.to_numpy()

test_X = test_X.to_numpy()
```

```python
print(train_X.shape)
print(train_y.shape)
print(test_X.shape)
```

```python
model = Sequential([
        Dense(units=25 , activation='relu'),
        Dense(units=15 , activation='relu'),
        Dense(units=1 , activation='linear')
    ])

model.compile(
    loss=BinaryCrossentropy(from_logits=True),
    optimizer=Adam(learning_rate=0.0001),
    metrics=['accuracy']
)
```

```python
model.fit(train_X, train_y, epochs=50)
```

```python
logit = model(test_X)

y_pred = tf.nn.sigmoid(logit)
```

```python
def mapping_results(x):
    if x >= 0.5:
        return 1
    else:
        return 0
```

```python
y_pred = tf.map_fn(mapping_results, y_pred)
```

```python
print(y_pred)
```

```python
survived = y_pred.numpy()
passenger_id = test_set['PassengerId'].to_numpy()

results = np.column_stack((passenger_id, survived))
```

```python
df = pd.DataFrame(results, columns=['PassengerId', 'Survived'])

df['PassengerId'] = df['PassengerId'].astype(int)
df['Survived'] = df['Survived'].astype(int)
```
